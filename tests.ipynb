```python
import pandas as pd

df = pd.read_csv("pipes.csv")
df
```

```python
df.plot.scatter(x="diameter", y="price");
```

```python
from sklearn import linear_model

# create model
model = linear_model.LinearRegression()
x = df[["diameter"]]
y = df[["price"]]

model.fit(x,y)
```

```python
print(model.predict([[62]])[0][0])
```
